```python
x1=dframe['Popularity']
```

```python
x1
```

```python
arr1=np.ravel(x1)# returns a 1d array
```

```python
arr1
```

```python
arr2=np.arange(len(arr1))
```

```python
sns.lineplot(x=arr2,y=arr1)
```

```python
#d=G*m+b where G is a 2x2 mqatrix , this is the vector matrix equation for y=m*x+b
G=np.stack([arr2,np.ones(arr2.shape)]).T
```

```python
G
```

```python
mhat,residuals,rank,s=np.linalg.lstsq(G,arr1)
```

```python
mhat
```

```python
X=np.linspace(0.0,4.0)
G1=np.stack([X,np.ones(X.shape)]).T
```

```python
print(X,G1)
```

```python
Y=np.dot(G1,mhat)
```

```python
plt.scatter(arr2,arr1)
sns.lineplot(x=arr2,y=arr1)
p=sns.lineplot(x=X,y=Y)
p.set(xlabel='serial no',ylabel='Popularity',title='Popularity least square fit')
```

```python
# Mean
m=song_df['Popularity'].mean()
print('The mean popularity is ',m)
```

```python
#standard deviation
import statistics as st
```

```python
sd=st.stdev(song_df['Popularity'])
print('The standard deviation in popularity is',sd)
```

```python
#variance
var=st.variance(song_df['Popularity'])
print('The variance in popularity is',var)
```

```python
#median
med=st.median(song_df['Popularity'])
print('The median popularity is',med)
```

```python
#mode
mod=st.mode(song_df['Popularity'])
print('The mode popularity is',mod)
```
